```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
country = spark.sparkContext.textFile("D:\School2\Dataset\GlobalLandTemperaturesByCountry.csv")
```

```python
# Extract the country field and count distinct countries
distinct_countries_count = country.filter(lambda line: "dt" not in line) \
    .map(lambda line: line.split(",")) \
    .map(lambda x: x[3]) \
    .distinct() \
    .count()

print(distinct_countries_count)
```

```python
# Find countries missing the data for the year 2011
country.filter(lambda line: "dt" not in line) \
    .map(lambda line: line.split(",")) \
    .map(lambda x: (x[0], x[1], x[2], x[3])) \
    .map(lambda x: (x[0].split("-")[0], x[0].split("-")[1], x[1], x[3])) \
    .map(lambda x: (x[0],x[1],x[2], x[3])) \
    .filter(lambda x: x[2] == '') \
    .filter(lambda x: x[0] == '2011') \
    .collect()
```

```python
# Filter out lines with "dt" and split the lines
country.filter(lambda line: "dt" not in line) \
    .map(lambda line: line.split(",")) \
    .map(lambda x: (x[0], x[1], x[2], x[3])) \
    .map(lambda x: (x[0].split("-")[0], x[1], x[3])) \
    .filter(lambda x: x[0] == '2011' and x[2] != 'Antarctica' ) \
    .map(lambda x: (x[2], (float(x[1]), 1))) \
    .reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1])) \
    .map(lambda x: (x[0], x[1][0] / x[1][1])) \
    .collect()
```
